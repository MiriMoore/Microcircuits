```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib
import numpy as np
```

```python
path = '/home/e/etayhay/frankm/Mazza2023/data/figures/Figure3/'

EEG_h = np.load('/home/e/etayhay/frankm/Mazza2023/data/simulations_processed/EEG_h.npy',allow_pickle=True)
EEG_m = np.load('/home/e/etayhay/frankm/Mazza2023/data/simulations_processed/EEG_m.npy',allow_pickle=True)

freqs = EEG_h.item(0)['mean_psd']['freq_wel']
single_fooof_freqs = np.load(path + 'single_fooof_freqs.npy')

EEG_bs = np.load(path + 'EEG_bs.npy')
aps_bs = np.load(path + 'aps_bs.npy')
ps_bs = np.load(path + 'ps_bs.npy')
```

```python
fig = plt.figure(figsize=(10,6))
labelfont = {'family':'arial','weight':'bold','size':14}

wrem = .1
hrem = .08
m_schem = fig.add_axes([0,.5,.5-wrem,.5-hrem],frameon=False)
absolute = fig.add_axes([.5,.5,.5-wrem,.5-hrem])
aperiodic = fig.add_axes([0,0,.5-wrem,.5-hrem],sharey=absolute)
periodic = fig.add_axes([.5,0,.5-wrem,.5-hrem])

m_schem.imshow(mpimg.imread(path + 'MDD_schematic.png'))
m_schem.set_ylim(970,85)
m_schem.set_xlim(0,880)
m_schem.set_xticks([])
m_schem.set_yticks([])

mddclr = 'red'#(0.8, 0.0, 0.45)

colors = ['k',mddclr]
inset0 = absolute.inset_axes([.6,.4,.4,.4])
inset1 = aperiodic.inset_axes([.6,.4,.4,.4])

for c in range(2):

    # previously did [c,0,low:high] to get 3:30
    absolute.plot(freqs,EEG_bs[c,0],color=colors[c],alpha=1)
    absolute.fill_between(freqs,EEG_bs[c,1,],EEG_bs[c,2],color = colors[c],alpha=.1)
    inset0.plot(freqs,EEG_bs[c,0],color=colors[c],alpha=1)
    inset0.fill_between(freqs,EEG_bs[c,1],EEG_bs[c,2],color = colors[c],alpha=.1)
    aperiodic.plot(single_fooof_freqs,aps_bs[c,0],color=colors[c],alpha=1)
                        
    aperiodic.fill_between(single_fooof_freqs,aps_bs[c,1],aps_bs[c,2],color = colors[c],alpha=.1)
    inset1.plot(single_fooof_freqs,aps_bs[c,0],color=colors[c],alpha=1)
    inset1.fill_between(single_fooof_freqs,aps_bs[c,1],aps_bs[c,2],color = colors[c],alpha=.1)

    periodic.plot(single_fooof_freqs,ps_bs[c,0],color=colors[c],alpha=1)
    periodic.fill_between(single_fooof_freqs,ps_bs[c,1],ps_bs[c,2],color = colors[c],alpha=.1)


insets = [inset0,inset1]
axes = [absolute, aperiodic,periodic]
[ax.set_xscale('log') for ax in insets]
[ax.set_yscale('log') for ax in insets] #this causes the weird 2 labels on the bootm
[ax.get_xaxis().set_major_formatter(matplotlib.ticker.ScalarFormatter()) for ax in insets]
[ax.get_yaxis().set_major_formatter(matplotlib.ticker.ScalarFormatter()) for ax in insets]

[ax.set_xlabel('Frequency (Hz)') for ax in axes[1:]]
absolute.set_ylabel(r'Power $\left( \frac{ V^{2} }{Hz} \right)$')
aperiodic.set_ylabel(r'Power $\left( \frac{ V^{2} }{Hz} \right)$')
periodic.set_ylabel(r'log(Power)')
 
[ax.set_xlim(4,30) for ax in axes]
[ax.set_xlim(4,30) for ax in insets]
[ax.set_xticks([4,10,30]) for ax in insets]
[ax.set_yticklabels('') for ax in insets]
inset1.set_ylim(inset0.get_ylim())

for ax in axes:
    leg = ax.legend(['Healthy','Depression'],fontsize=12,frameon=False,loc='upper right', bbox_to_anchor=(1.08,1.1))
    for line in leg.get_lines():
        line.set_linewidth(3.0)


[ax.spines['right'].set_visible(False) for ax in axes]
[ax.spines['top'].set_visible(False) for ax in axes]
[ax.spines['right'].set_visible(False) for ax in insets]
[ax.spines['top'].set_visible(False) for ax in insets]

labels = [r'$\theta$',r'$\alpha$',r'$\beta$']
[ax.axvline(i, ls='--',alpha=.4, color = 'k') for i in [4,8,12] for ax in axes]
[ax.axvline(i, ls='--',alpha=.4, color = 'k') for i in [4,8,12] for ax in insets]
[ax.annotate(xy=(x,1e-16),s=t,fontsize=10) for x,t in zip([5,8.5,22],labels) for ax in insets]


[ax.annotate(xy=(x,0),s=t,fontsize=10) for x,t in zip([5.5,9.5,22],labels) for ax in [absolute,aperiodic]]

periodic.set_ylim(-.08,.65)
[periodic.annotate(xy=(x,-.05),s=t,fontsize=10) for x,t in zip([5.5,9.5,22],labels)]

fig.text(x=-.065,y=.93,s='A',**labelfont)
absolute.set_title('B',loc='left', y=1, x=-.17,**labelfont)
aperiodic.set_title('C',loc='left', y=1, x=-.17,**labelfont)
periodic.set_title('D',loc='left', y=1, x=-.17,**labelfont)

inset0.tick_params(axis = 'x',which = 'both',labelbottom=False)
inset1.tick_params(axis = 'x',which = 'both',labelbottom=False)

fig.text(x=.753,y=.633,s='5')
fig.text(x=.804,y=.633,s='10')
fig.text(x=.89,y=.633,s='30')

xshift = -.5
yshift = -.5
fig.text(x=.753+xshift,y=.633+yshift,s='5')
fig.text(x=.804+xshift,y=.633+yshift,s='10')
fig.text(x=.89+xshift,y=.633+yshift,s='30')

l, h = 9e-16,4.2e-14
inset0.set_ylim(l,h)
inset1.set_ylim(l,h)

periodic.set_ylim(0,.55)

fig.savefig('/home/e/etayhay/frankm/Mazza2023/figures/figoutput/Figure3.jpeg', facecolor='white', edgecolor='none',bbox_inches = "tight",dpi=600)
```
