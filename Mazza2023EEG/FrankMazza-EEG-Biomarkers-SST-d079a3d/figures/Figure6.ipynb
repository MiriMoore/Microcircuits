```python
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.collections import LineCollection
from matplotlib.collections import PolyCollection
import matplotlib.image as img
import matplotlib.patches as mpatches
from mpl_toolkits.axes_grid1.anchored_artists import AnchoredSizeBar
from mpl_toolkits.axes_grid.inset_locator import (inset_axes, InsetPosition,mark_inset)
import numpy as np
np.seterr(divide='ignore', invalid='ignore')
from PIL import Image
```

```python
path = '/home/e/etayhay/frankm/Mazza2023/data/figures/Figure6/'
dipole = plt.imread(path + 'Figure6_A.jpg')
B_img = plt.imread(path + 'Figure6_B.jpg')
C_img = plt.imread(path + 'Figure6_C.jpg')
D_img = plt.imread(path + 'Figure6_theta.jpg')
E_img = plt.imread(path + 'Figure6_alpha.jpg')
F_img = plt.imread(path + 'Figure6_beta.jpg')
```

```python
fig = plt.figure(figsize=(10,10))

frameon = True

ytop = .38
A = fig.add_axes([0,ytop,.3,.3], frameon=frameon)
B = fig.add_axes([.33,ytop,.3,.3],frameon=frameon)
C = fig.add_axes([.66,ytop,.3,.3],frameon=frameon)
D = fig.add_axes([0,0,.3,.3],frameon=frameon)
E = fig.add_axes([.33,0,.3,.3],frameon=frameon)
F = fig.add_axes([.66,0,.3,.3],frameon=frameon)

BC_colorbar = fig.add_axes([.98,ytop,.1,.3],frameon=frameon)
DEF_colorbar = fig.add_axes([.98,0,.1,.3],frameon=frameon)

BC_colorbar.axis('off')
DEF_colorbar.axis('off')

A.imshow(dipole)
A.set_xlim(800,7200)
A.set_ylim(8350,1600)


topos = [B_img,C_img, D_img,E_img,F_img]
axes = [B,C,D,E,F]

for ax,topo in list(zip(axes,topos)):
    ax.imshow(topo)
    ax.set_xlim(0,4300)
    ax.set_ylim(7000,2780)

cbars = [C_img,C_img]
axes = [BC_colorbar,DEF_colorbar]

for ax,cbar in list(zip(axes,cbars)):
    ax.imshow(cbar)
    ax.set_xlim(5100,9100)
    ax.set_ylim(10000,270)



axes = [A,B,C,D,E,F]

[ax.axis('off') for ax in axes] # remove ticks after formatting

# # need to place manually since fig dims change with image
y0,y1 = .68,.3
x1,x2,x3 = 0,.33,.66

fig.text(x=x1,y=y0+.01,s='A',weight='bold',fontsize=12)
fig.text(x=x2,y=y0+.01,s='B',weight='bold',fontsize=12)
fig.text(x=x3,y=y0+.01,s='C',weight='bold',fontsize=12)
fig.text(x=x1,y=y1+.01,s='D',weight='bold',fontsize=12)
fig.text(x=x2,y=y1+.01,s='E',weight='bold',fontsize=12)
fig.text(x=x3,y=y1+.01,s='F',weight='bold',fontsize=12)

# Other titles
xshift = .13
fig.text(x=x1+xshift,y=y1+.01,s='Theta',fontsize=12)
fig.text(x=x2+xshift,y=y1+.01,s='Alpha',fontsize=12)
fig.text(x=x3+xshift,y=y1+.01,s='Beta',fontsize=12)

# Cbar titles
xshift = .36
yshift = -.02
fig.text(x=x3+xshift,y=y0+yshift,s=u'\u03bc'+'V',fontsize=12) #top
fig.text(x=x3+xshift-.037,y=y1+yshift+.005,s=r'Power $\left( \frac{ \mu V^{2} }{Hz} \right)$',fontsize=12) #bottom

# Cbar ticks & labels
# Top
xtickshift = .059
ytickshift = .028
ticks = np.linspace(ytop+.002,y0+yshift-.044,9)
for tick in ticks:
    fig.text(x=x3+xshift+xtickshift,y=tick+ytickshift,s='-',fontsize=12)
    
labels = ['- 0.4', '- 0.3','- 0.2','- 0.1','0','0.1','0.2','0.3','0.4']
for t in range(len(ticks)):
    fig.text(x=x3+xshift+xtickshift+.014,y=ticks[t]+ytickshift,s=labels[t],fontsize=8)
    
# Bottom
xtickshift = .059
ytickshift = .028
ticks = np.linspace(0.002,y1+yshift-.044,9)
for tick in ticks:
    fig.text(x=x3+xshift+xtickshift,y=tick+ytickshift,s='-',fontsize=12)

labels = ['- 0.4', '- 0.3','- 0.2','- 0.1','0','0.1','0.2','0.3','0.4']
for t in range(len(ticks)):
    fig.text(x=x3+xshift+xtickshift+.014,y=ticks[t]+ytickshift,s=labels[t],fontsize=8)
    
fig.savefig('/home/e/etayhay/frankm/Mazza2023/figures/figoutput/Figure6.jpeg', facecolor='white', edgecolor='none',bbox_inches = "tight",dpi=600)
```
