```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib
import numpy as np

path = '/home/e/etayhay/frankm/Mazza2023/data/figures/Figure4/'

EEG_h = np.load('/home/e/etayhay/frankm/Mazza2023/data/simulations_processed/EEG_h.npy',allow_pickle=True)
EEG_m = np.load('/home/e/etayhay/frankm/Mazza2023/data/simulations_processed/EEG_m.npy',allow_pickle=True)

freqs = EEG_h.item(0)['mean_psd']['freq_wel']
single_fooof_freqs = np.load(path + 'single_fooof_freqs.npy')

EEG_bs = np.load(path + 'EEG_bs.npy')
aps_bs = np.load(path + 'aps_bs.npy')
ps_bs = np.load(path + 'ps_bs.npy')
```

```python
fig = plt.figure(figsize = (12,5))
w = .26
h = .4

absolute_PV = fig.add_axes([0,0,w,h])
aperiodic_PV = fig.add_axes([.33,0,w,h])
periodic_PV = fig.add_axes([.66,0,w,h])

absolute_SST = fig.add_axes([0,.65,w,h])
aperiodic_SST = fig.add_axes([.33,.65,w,h])
periodic_SST = fig.add_axes([.66,.65,w,h])

ap_inset_SST = aperiodic_SST.inset_axes([.5,.5,.5,.5])
abs_inset_SST = absolute_SST.inset_axes([.5,.5,.5,.5])

ap_inset_PV = aperiodic_PV.inset_axes([.5,.5,.5,.5])
abs_inset_PV = absolute_PV.inset_axes([.5,.5,.5,.5])

labelfont = {'family':'arial','weight':'bold','size':14}
leglabels = [None,'20%','40%','60%']

axes = [[absolute_SST, aperiodic_SST, periodic_SST],[absolute_PV,aperiodic_PV,periodic_PV]]

colors = [['black','brown','red','orange'], ['black','cadetblue','green','limegreen']]

insets = [[abs_inset_SST,ap_inset_SST],[abs_inset_PV,ap_inset_PV]]

for i in range(2):
    for d in range(4):
        axes[i][0].plot(freqs,EEG_bs[i][d][0], color=colors[i][d],label=leglabels[d])
        axes[i][0].fill_between(freqs,EEG_bs[i][d][1],EEG_bs[i][d][2], color=colors[i][d], alpha=.2)
                        
        insets[i][0].plot(freqs,EEG_bs[i][d][0], color=colors[i][d],label=leglabels[d])
        insets[i][0].fill_between(freqs,EEG_bs[i][d][1],EEG_bs[i][d][2], color=colors[i][d], alpha=.2)
        
        axes[i][1].plot(single_fooof_freqs,aps_bs[i][d][0],color=colors[i][d],label=leglabels[d])
        axes[i][1].fill_between(single_fooof_freqs,aps_bs[i][d][1],aps_bs[i][d][2], color=colors[i][d], alpha=.2)
        
        insets[i][1].plot(single_fooof_freqs,aps_bs[i][d][0],color=colors[i][d],label=leglabels[d])
        insets[i][1].fill_between(single_fooof_freqs,aps_bs[i][d][1],aps_bs[i][d][2], color=colors[i][d], alpha=.2)
        
        
        axes[i][2].plot(single_fooof_freqs,ps_bs[i][d][0],color=colors[i][d],label=leglabels[d])
        axes[i][2].fill_between(single_fooof_freqs,ps_bs[i][d][1],ps_bs[i][d][2], color=colors[i][d], alpha=.2)
        
insets = [abs_inset_SST,ap_inset_SST,abs_inset_PV,ap_inset_PV]
        
[[inset.spines[spine].set_visible(False) for spine in ['top','right']] for inset in insets]
[inset.set_xscale('log') for inset in insets]
[inset.set_yscale('log') for inset in insets]
[inset.get_xaxis().set_major_formatter(matplotlib.ticker.ScalarFormatter()) for inset in insets]
[inset.get_yaxis().set_major_formatter(matplotlib.ticker.ScalarFormatter()) for inset in insets]
[inset.set_xlim(4,30) for inset in insets]
[inset.set_xticks([4,10,30]) for inset in insets]
[inset.set_xticklabels('') for inset in insets]
[inset.set_yticklabels('') for inset in insets]
# ap_inset_SST.set_ylim(abs_inset_SST.get_ylim())
# ap_inset_PV.set_ylim(abs_inset_PV.get_ylim())

axes = [absolute_SST, aperiodic_SST,periodic_SST,absolute_PV, aperiodic_PV,periodic_PV]

[ax.set_xlabel('Frequency (Hz)') for ax in axes]
[ax.set_xlim(4,30) for ax in axes]
[ax.spines['right'].set_visible(False) for ax in axes]
[ax.spines['top'].set_visible(False) for ax in axes]

labels = [r'$\theta$',r'$\alpha$',r'$\beta$']
[ax.axvline(i, ls='--',alpha=.4, color = 'k') for i in [4,8,12] for ax in axes]
[ax.axvline(i, ls='--',alpha=.4, color = 'k') for i in [4,8,12] for ax in insets]
#[ax.annotate(xy=(x,1e-15),s=t,fontsize=10) for x,t in zip([5,8.5,22],labels) for ax in insets]

absolute_SST.set_ylim(-.37e-14,absolute_SST.get_ylim()[1])
aperiodic_SST.set_ylim(-.37e-14,absolute_SST.get_ylim()[1])
periodic_SST.set_ylim(-.08,.9)

absolute_PV.set_ylim(-.37e-14,absolute_PV.get_ylim()[1])
aperiodic_PV.set_ylim(-.37e-14,absolute_PV.get_ylim()[1])
periodic_PV.set_ylim(-.08,.9)

[ax.annotate(xy=(x,-.2e-14),s=t,fontsize=10) for x,t in zip([5.5,9.5,22],labels) for ax in [absolute_SST,aperiodic_SST,absolute_PV,aperiodic_PV]]
[periodic_SST.annotate(xy=(x,-.05),s=t,fontsize=10) for x,t in zip([5.5,9.5,22],labels)]
[periodic_PV.annotate(xy=(x,-.05),s=t,fontsize=10) for x,t in zip([5.5,9.5,22],labels)]

absolute_SST.set_ylabel(r'Power $\left( \frac{ V^{2} }{Hz} \right)$')
aperiodic_SST.set_ylabel(r'Power $\left( \frac{ V^{2} }{Hz} \right)$')
periodic_SST.set_ylabel(r'log(Power)')
absolute_PV.set_ylabel(r'Power $\left( \frac{ V^{2} }{Hz} \right)$')
aperiodic_PV.set_ylabel(r'Power $\left( \frac{ V^{2} }{Hz} \right)$')
periodic_PV.set_ylabel(r'log(Power)')

absolute_SST.set_title('A',loc='left', y=1, x=-.17, **labelfont)
aperiodic_SST.set_title('B',loc='left', y=1, x=-.17, **labelfont)
periodic_SST.set_title('C',loc='left', y=1, x=-.17, **labelfont)
absolute_PV.set_title('D',loc='left', y=1, x=-.17, **labelfont)
aperiodic_PV.set_title('E',loc='left', y=1, x=-.17, **labelfont)
periodic_PV.set_title('F',loc='left', y=1, x=-.17, **labelfont)


leg = aperiodic_SST.legend(ncol=3, fontsize=12,frameon=False,loc='upper right', bbox_to_anchor=(1.06,1.28))
for line in leg.get_lines():
    line.set_linewidth(3.0)

leg = aperiodic_PV.legend(ncol=3, fontsize=12,frameon=False,loc='upper right', bbox_to_anchor=(1.06,1.28))
for line in leg.get_lines():
    line.set_linewidth(3.0)
    
    
ap_inset_SST.tick_params(axis = 'x',which = 'both',labelbottom=False)
abs_inset_SST.tick_params(axis = 'x',which = 'both',labelbottom=False)
ap_inset_PV.tick_params(axis = 'x',which = 'both',labelbottom=False)
abs_inset_PV.tick_params(axis = 'x',which = 'both',labelbottom=False)

for xshift,yshift in zip([-.025,.305,-.025,.305],[.045,.045,.695,.695]):
    fig.text(x=.152+xshift,y=.115+yshift,s='4')
    fig.text(x=.21+xshift,y=.115+yshift,s='10')
    fig.text(x=.275+xshift,y=.115+yshift,s='30')


fig.text(x=.31, y=1.15, s= 'Reduced SST Interneuron Inhibition', fontsize=14)
fig.text(x=.31, y=.5,s='Reduced PV Interneuron Inhibition', fontsize=14)

l, h = 9e-16,4.2e-14
abs_inset_SST.set_ylim(l,h)
ap_inset_SST.set_ylim(l,h)

l, h = 9e-16,2.2e-14
abs_inset_PV.set_ylim(l,h)
ap_inset_PV.set_ylim(l,h)

fig.savefig('/home/e/etayhay/frankm/Mazza2023/figures/figoutput/Figure4.jpeg', facecolor='white', edgecolor='none',bbox_inches = "tight",dpi=600)
```
